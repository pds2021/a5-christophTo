## App - Decidual Trees Classifier

When you go for a walk in the forest again and have to listen to your parents saying that today's young people have no idea about their environment and nature. Take a picture of a leaf an upload it and get the Latin and German name of the tree. So you can receive enthusiasm even from older generations without having memorized the trees.
(IMPORTANT: this only handles deciduous trees - Beech, Oak, Alder, Chestnut, Birch. These cover 58% of the trees in Germany.


```python
# imports
#!pip install -Uqq fastai
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```python
download_url('https://github.com/pds2021/a5-christophTo/releases/download/Solutions/export.pkl', 'tree_classifier.pkl')
learn_inf = load_learner('tree_classifier.pkl', cpu=True)
btn_upload = widgets.FileUpload()
btn_run = widgets.Button(description='Classify')
out_pl = widgets.Output()
lbl_pred = widgets.Label()

def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(512,512))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
VBox([widgets.Label('Select your Leaf!'), 
      btn_upload, btn_run, out_pl, lbl_pred])
```
