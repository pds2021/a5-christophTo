## App - Decidual Trees Classifier

When you go for a walk in the forest again and have to listen to your parents saying that today's young people have no idea about their environment and nature. Take a picture of a leaf an upload it and get the Latin and English name of the tree. So you can receive enthusiasm even from older generations without having memorized the trees.
(IMPORTANT: this only handles deciduous trees - Maple, Birch, Beech, Oak, Alder. These cover 58% of the trees in Germany.


```python
# imports
#!pip install -Uqq fastai
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```python
download_url('https://github.com/pds2021/a5-christophTo/releases/download/Solutions/export.pkl', 'tree_classifier.pkl')
learn_inf = load_learner('tree_classifier.pkl', cpu=True)
btn_upload = widgets.FileUpload()
btn_upload.style.button_color = 'lightblue'
btn_run = widgets.Button(description='Classify')
btn_run.style.button_color = 'lightblue'
out_pl = widgets.Output()
lbl_pred = widgets.Label(value = "Results:")
lbl_pred_One = widgets.Label(value = "Tree: ???")
lbl_pred_Two = widgets.Label(value = "Probability: ???")
lbl_pred_Three = widgets.Label(value = "Maximum height: ???")
lbl_pred_Four = widgets.Label(value = "Maximum age: ???")
lbl_pred_Five = widgets.Label(value = "Share in Germany: ??%")

def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(512,512))
    pred,pred_idx,probs = learn_inf.predict(img)
    height = "test"
    age = "test"
    percentage = "test"
    if pred == "acer pseudoplatanus":
      pred = "Maple (Acer pseudoplatanus)"
      height = "30 m"
      age = "500 years"
      percentage = "6%"
    if pred == "betula pendula":
      pred = "Birch (Betula pendula)"
      height = "28 m"
      age = "120 years"
      percentage = "10%"
    if pred == "fagus sylvatica":
      pred = "Beech (Fagus sylvatica)"
      height = "45 m"
      age = "300 years"
      percentage = "16%"
    if pred == "quercus robur":
      pred = "Oak (Quercus robur)"
      height = "40 m"
      age = "1000 years"
      percentage = "10%"
    if pred == "alnus glutinosa":
      pred = "Alder (Alnus glutinosa)"
      height = "28 m"
      age = "120 years"
      percentage = "10%"
    if pred == "aesculus hippocastanum":
      pred = "Chestnut (Aesculus hippocastanum)"
      height = "25 m"
      age = "200 years"
      percentage = "6%"
    lbl_pred.value = f'Results:'
    lbl_pred_One.value = f'Tree: {pred}'
    lbl_pred_Two.value = f'Probability: {probs[pred_idx]:.04f}'
    lbl_pred_Three.value = f'Maximum height: {height}'
    lbl_pred_Four.value = f'Maximum age: {age}'
    lbl_pred_Five.value = f'Share in Germany: {percentage}'

btn_run.on_click(on_click_classify)
left_box = VBox([widgets.Label('Select your image for classification!'), btn_upload, btn_run, out_pl])
right_box = VBox([lbl_pred, lbl_pred_One, lbl_pred_Two, lbl_pred_Three, lbl_pred_Four, lbl_pred_Five])
HBox([left_box, right_box])
```
